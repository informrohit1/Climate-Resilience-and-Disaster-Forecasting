```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("final_data.csv")
df.head(1)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'],format='mixed')
```

```python
df.drop(['lat','lng','created_time'],axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.replace('Null', np.nan, inplace=True)
```

```python
df[['temperature', 'max_temperature', 'humidity', 'wind_speed', 'precipitation', 'uv_index']] = df[['temperature', 'max_temperature', 'humidity', 'wind_speed', 'precipitation', 'uv_index']].astype('float64')
```

```python
df['continent'].unique()
```

```python
nar = df[df['continent']=='nar']
asia = df[df['continent']=='asia']
sar = df[df['continent']=='sar']
ant = df[df['continent']=='ant']
afr = df[df['continent']=='afr']
eur = df[df['continent']=='eur']
aus = df[df['continent']=='aus']
ocean = df[df['continent']=='ocean']
```

```python
nar
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
nar[cols] = nar[cols].apply(pd.to_numeric, errors='coerce')
nar[cols] = nar[cols].fillna(nar[cols].mean())
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
asia[cols] = asia[cols].apply(pd.to_numeric, errors='coerce')
asia[cols] = asia[cols].fillna(asia[cols].mean())
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
sar[cols] = sar[cols].apply(pd.to_numeric, errors='coerce')
sar[cols] = sar[cols].fillna(sar[cols].mean())
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
ant[cols] = ant[cols].apply(pd.to_numeric, errors='coerce')
ant[cols] = ant[cols].fillna(ant[cols].mean())
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
afr[cols] = afr[cols].apply(pd.to_numeric, errors='coerce')
afr[cols] = afr[cols].fillna(afr[cols].mean())
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
eur[cols] = eur[cols].apply(pd.to_numeric, errors='coerce')
eur[cols] = eur[cols].fillna(eur[cols].mean())
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
aus[cols] = aus[cols].apply(pd.to_numeric, errors='coerce')
aus[cols] = aus[cols].fillna(aus[cols].mean())
```

```python
cols=['temperature','max_temperature','min_temperature','humidity','wind_speed','precipitation','uv_index']
ocean[cols] = ocean[cols].apply(pd.to_numeric, errors='coerce')
ocean[cols] = ocean[cols].fillna(ocean[cols].mean())
```

```python
result = pd.concat([nar,asia,sar,afr,ant,eur,aus,ocean], axis=0)
```
