```python
import http.client
import json
import pandas as pd

headers = {
    'x-api-key': "08af8b0cedf4f975507cb122476f0d32846546b9a06b4b5688e2632197afb32c",
    'Content-type': "application/json"
}

all_data = []

for i in range(1, 10):
    conn = http.client.HTTPSConnection("api.ambeedata.com")
    conn.request("GET", f"/disasters/latest/by-continent?continent=NAR&limit=50&page={i}", headers=headers)
    
    res = conn.getresponse()
    data = res.read()
    data = data.decode("utf-8")
    json_data = json.loads(data)

    if 'result' in json_data:
        all_data.extend(json_data['result'])
    
    conn.close()

# Create DataFrame from the collected data
df = pd.DataFrame(all_data)[['event_type', 'event_name', 'date', 'lat', 'lng', 'continent', 'country_code', 'created_time']]
df
```
