```python
import pandas as pd
import requests

def get_weather_data(api_key, latitude, longitude, date):
    url = f"http://api.weatherapi.com/v1/history.json?key={api_key}&q={latitude},{longitude}&dt={date}"
    response = requests.get(url)

    if response.status_code == 200:
        weather_data = response.json()
        
        if 'forecast' in weather_data:
            day_data = weather_data['forecast']['forecastday'][0]['day']
            weather_condition = day_data.get('condition', {}).get('text', "Null")  # Using .get() for safety

            try:
                temperature = day_data.get('avgtemp_c', "Null")
                max_temp = day_data.get('maxtemp_c', "Null")
                min_temp = day_data.get('mintemp_c', "Null")
                humidity = day_data.get('avghumidity', "Null")
                wind_speed = day_data.get('maxwind_kph', "Null")
                precipitation = day_data.get('totalprecip_mm', "Null")
                cloud_cover = day_data.get('cloud', "Null")
                uv_index = day_data.get('uv', "Null")
                visibility = weather_data.get('current', {}).get('vis_km', "Null")  # Current visibility

                return {
                    'temperature': temperature,
                    'max_temperature': max_temp,
                    'min_temperature': min_temp,
                    'humidity': humidity,
                    'wind_speed': wind_speed,
                    'precipitation': precipitation,
                    'cloud_cover': cloud_cover,
                    'uv_index': uv_index,
                    'condition': weather_condition,
                    'visibility': visibility
                }
            except Exception as e:
                print(f"Error processing data for {date} at ({latitude}, {longitude}): {e}")
                return {
                    'temperature': "Null",
                    'max_temperature': "Null",
                    'min_temperature': "Null",
                    'humidity': "Null",
                    'wind_speed': "Null",
                    'precipitation': "Null",
                    'cloud_cover': "Null",
                    'uv_index': "Null",
                    'condition': weather_condition,
                    'visibility': "Null"
                }
        else:
            return {
                'temperature': "Null",
                'max_temperature': "Null",
                'min_temperature': "Null",
                'humidity': "Null",
                'wind_speed': "Null",
                'precipitation': "Null",
                'cloud_cover': "Null",
                'uv_index': "Null",
                'condition': "Null",
                'visibility': "Null"
            }
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return {
            'temperature': "Null",
            'max_temperature': "Null",
            'min_temperature': "Null",
            'humidity': "Null",
            'wind_speed': "Null",
            'precipitation': "Null",
            'cloud_cover': "Null",
            'uv_index': "Null",
            'condition': "Null",
            'visibility': "Null"
        }

# Example usage
if __name__ == "__main__":
    api_key = 'bf320090dd5f4c30a67201629242609'  # Replace with your WeatherAPI key


    results = []

    for index, row in data.iterrows():
        date = row['date']
        latitude = row['lat']
        longitude = row['lng']
        
        weather = get_weather_data(api_key, latitude, longitude, date)
        
        results.append({
            'date': date,
            'latitude': latitude,
            'longitude': longitude,
            **weather
        })

    results_df = pd.DataFrame(results)
    print(results_df)
```
